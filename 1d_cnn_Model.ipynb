```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, roc_auc_score
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Conv1D, MaxPooling1D, Flatten, Input
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# CSV 파일 로드
file_path = 'combined_dataset_with_multi_labels.csv'
df = pd.read_csv(file_path)

# 상관계수 0.3 이상인 피처 리스트 (기존과 동일)
selected_features = [
    'Flow Duration', 'Fwd IAT Total', 'Bwd IAT Total', 'Fwd IAT Max',
    'Bwd IAT Max', 'RST Flag Count', 'Bwd IAT Mean', 'Fwd IAT Mean',
    'Bwd IAT Min', 'Fwd PSH Flags', 'Fwd IAT Min', 'FIN Flag Count',
    'Fwd IAT Std', 'Packet Length Std', 'Bwd Packet Length Std',
    'Average Packet Size', 'Packet Length Max', 'Bwd IAT Std',
    'Packet Length Mean', 'Bwd Packet Length Max', 'Dst Port'
]

# 데이터프레임에서 피처 선택
df_selected = df[selected_features]

# 로그 변환 적용 (log1p)
df_log_transformed = df_selected.copy()
for feature in selected_features:
    df_log_transformed[feature] = np.log1p(df_log_transformed[feature])

# Label에 따라 정상(0)과 비정상(>=1)으로 분류
df['Is Malicious'] = df['Label'].apply(lambda x: 1 if x >= 1 else 0)

# 양성 클래스(비정상) 데이터를 0.3%만 샘플링
positive_df = df_log_transformed[df['Is Malicious'] == 1]
negative_df = df_log_transformed[df['Is Malicious'] == 0]
positive_labels = df[df['Is Malicious'] == 1]['Is Malicious']
negative_labels = df[df['Is Malicious'] == 0]['Is Malicious']

# 양성 클래스에서 0.3%만 무작위 샘플링
sample_frac = 0.3
positive_sampled_df = positive_df.sample(frac=sample_frac, random_state=42)
positive_sampled_labels = positive_labels.loc[positive_sampled_df.index]

# 샘플링된 양성 클래스와 음성 클래스 데이터를 결합
X = pd.concat([positive_sampled_df, negative_df], axis=0)
y = pd.concat([positive_sampled_labels, negative_labels], axis=0)

# 샘플링 후 데이터 분포 확인
print("샘플링 후 데이터 분포:")
print(y.value_counts())

# SMOTE를 사용하여 오버샘플링
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# 오버샘플링 후 데이터 분포 확인
print("\nSMOTE 적용 후 데이터 분포:")
print(pd.Series(y_resampled).value_counts())

# Z-score 정규화 적용 (StandardScaler)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_resampled)

# 오토인코더 모델 정의 함수
def build_autoencoder(input_dim, encoding_dim):
    input_layer = Input(shape=(input_dim,))
    # Encoder
    encoded = Dense(16, activation='relu')(input_layer)
    encoded = Dense(encoding_dim, activation='relu')(encoded)
    # Decoder
    decoded = Dense(16, activation='relu')(encoded)
    decoded = Dense(input_dim, activation='linear')(decoded)  # sigmoid 대신 linear 사용 가능
    autoencoder = Model(input_layer, decoded)
    encoder = Model(input_layer, encoded)
    autoencoder.compile(optimizer='adam', loss='mse')
    return autoencoder, encoder


# 오토인코더 생성 (입력 차원과 축소 차원 설정)
input_dim = X_scaled.shape[1]  # 입력 데이터 차원
encoding_dim = 10  # 축소된 차원의 갯수 지정
autoencoder, encoder = build_autoencoder(input_dim, encoding_dim)

# 오토인코더 학습
history_ae = autoencoder.fit(X_scaled, X_scaled, epochs=50, batch_size=32, validation_split=0.2, verbose=1)

# 인코더를 사용해 차원 축소
X_encoded = encoder.predict(X_scaled)

# 1D CNN 입력 형태로 reshape
X_encoded_cnn = X_encoded.reshape((X_encoded.shape[0], X_encoded.shape[1], 1))

# 학습 데이터와 테스트 데이터 분리 (오토인코더로 축소된 데이터 사용)
X_train, X_test, y_train, y_test = train_test_split(X_encoded_cnn, y_resampled, test_size=0.3, random_state=42)

# Keras 1D CNN 모델 정의
input_shape = (X_train.shape[1], 1)  # (timesteps, channels)
keras_model = Sequential([
    Input(shape=input_shape),
    Conv1D(filters=32, kernel_size=3, activation='gelu', padding='same'),
    MaxPooling1D(pool_size=2),
    Conv1D(filters=64, kernel_size=3, activation='gelu', padding='same'),
    MaxPooling1D(pool_size=2),
    Flatten(),
    Dense(50, activation='gelu'),
    Dense(1, activation='sigmoid')
])

# 모델 컴파일
keras_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
)

# 콜백 설정
early_stopping = EarlyStopping(
    monitor='val_auc',
    patience=20,
    mode='max',
    restore_best_weights=True
)
reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.2,
    patience=10,
    min_lr=0.0001
)

# 모델 학습
history = keras_model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=200,
    batch_size=32,
    callbacks=[early_stopping, reduce_lr],
    verbose=1
)

# 예측 (원본 Keras 모델)
y_pred_proba = keras_model.predict(X_test).flatten()
y_pred = (y_pred_proba >= 0.5).astype(int)

# 모델 성능 평가 (원본 모델)
print("\nOriginal Keras 1D CNN Model Classification Report:")
print(classification_report(y_test, y_pred, target_names=['Normal', 'Malicious']))
roc_auc = roc_auc_score(y_test, y_pred_proba)
print(f"Original Keras 1D CNN Model ROC-AUC Score: {roc_auc:.4f}")

# 학습 곡선 시각화
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history.history['auc'], label='Train AUC')
plt.plot(history.history['val_auc'], label='Validation AUC')
plt.title('AUC Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('AUC')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.savefig('training_history_with_autoencoder_cnn1d.png')
plt.show()

# Keras 모델 저장
keras_model.save('cnn1d_model_with_autoencoder_keras.keras')

# TFLite 모델 양자화 및 성능 평가 (16비트 양자화 제외)
quantization_levels = {
    'dynamic': None,  # 동적 양자화 (가중치만 양자화, 활성화는 float32)
    'int8': tf.int8   # 완전 정수 양자화
}

results = {}

def representative_dataset():
    for i in range(200):
        yield [X_train[i:i+1].astype(np.float32)]

for level_name, dtype in quantization_levels.items():
    print(f"\nQuantizing model to {level_name}...")
    converter = tf.lite.TFLiteConverter.from_keras_model(keras_model)
    converter.optimizations = [tf.lite.Optimize.DEFAULT]
    
    if level_name == 'dynamic':
        pass  # 기본 설정으로 충분
    else:
        # 정수 양자화: 대표 데이터셋 필요
        converter.target_spec.supported_types = [dtype]
        converter.inference_input_type = dtype
        converter.inference_output_type = dtype
        converter.representative_dataset = representative_dataset
    
    tflite_model = converter.convert()
    
    # TFLite 모델 저장
    tflite_filename = f'cnn1d_quantized_with_autoencoder_{level_name}_keras.tflite'
    with open(tflite_filename, 'wb') as f:
        f.write(tflite_model)
    print(f"Model quantized and saved as '{tflite_filename}'")
    
    # TFLite 모델 로드 및 예측
    interpreter = tf.lite.Interpreter(model_content=tflite_model)
    interpreter.allocate_tensors()
    
    input_details = interpreter.get_input_details()
    output_details = interpreter.get_output_details()
    
    y_pred_proba_tflite = []
    for i in range(len(X_test)):
        input_data = X_test[i:i+1].astype(np.float32)
        if level_name != 'dynamic' and dtype in [tf.int8]:
            # 정수 양자화 모델의 경우 입력 데이터를 양자화
            input_scale, input_zero_point = input_details[0]['quantization']
            input_data = (input_data / input_scale + input_zero_point).astype(np.int8)
        
        interpreter.set_tensor(input_details[0]['index'], input_data)
        interpreter.invoke()
        output_data = interpreter.get_tensor(output_details[0]['index'])
        
        if level_name != 'dynamic' and dtype in [tf.int8]:
            # 정수 양자화 모델의 출력 역변환
            output_scale, output_zero_point = output_details[0]['quantization']
            output_data = (output_data.astype(np.float32) - output_zero_point) * output_scale
        else:
            output_data = output_data.astype(np.float32)
        
        y_pred_proba_tflite.append(output_data[0])
    
    y_pred_proba_tflite = np.array(y_pred_proba_tflite).flatten()
    y_pred_tflite = (y_pred_proba_tflite >= 0.5).astype(int)
    
    # 성능 평가
    print(f"\nClassification Report for {level_name} Quantized Model:")
    print(classification_report(y_test, y_pred_tflite, target_names=['Normal', 'Malicious']))
    roc_auc_tflite = roc_auc_score(y_test, y_pred_proba_tflite)
    print(f"ROC-AUC Score for {level_name} Quantized Model: {roc_auc_tflite:.4f}")
    
    results[level_name] = {
        'roc_auc': roc_auc_tflite,
        'classification_report': classification_report(y_test, y_pred_tflite, output_dict=True)
    }

# 결과 요약 출력
print("\nSummary of Quantization Results:")
for level_name, result in results.items():
    print(f"{level_name}: ROC-AUC = {result['roc_auc']:.4f}")

# 스케일러와 오토인코더 객체 저장
scaler_filename = 'scaler_with_autoencoder_cnn1d.pkl'
joblib.dump(scaler, scaler_filename)
print(f"Scaler saved as {scaler_filename}")

encoder_filename = 'encoder_with_autoencoder_cnn1d.keras'
encoder.save(encoder_filename)
print(f"Encoder saved as {encoder_filename}")
```
