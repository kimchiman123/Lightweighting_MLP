```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, roc_auc_score
from imblearn.over_sampling import SMOTE
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import tensorflow_model_optimization as tfmot
import tensorflow_addons as tfa # Note: tensorflow_addons is not explicitly used, but kept as it was in the original import

# 시각화 함수

def plot_training_history(history, title_prefix, save_filename):
    """
    history로부터 학습 및 검증 손실과 AUC를 그래프로 표시합니다.
    일반 모델과 증류 모델의 서로 다른 손실 키를 처리합니다.
    """
    # 손실에 다라 키 설정 
    if 'student_loss' in history.history:
        loss_key = 'student_loss'
        val_loss_key = 'val_student_loss'
        loss_label = 'Train Loss (Student)'
        val_loss_label = 'Validation Loss (Student)'
    else:
        loss_key = 'loss'
        val_loss_key = 'val_loss'
        loss_label = 'Train Loss'
        val_loss_label = 'Validation Loss'

    # AUC가 히스토리에 있는지 확인
    auc_key = 'auc'
    val_auc_key = 'val_auc'

    plt.figure(figsize=(12, 4))

    # 서브플롯 1: 손실
    plt.subplot(1, 2, 1)
    if loss_key in history.history:
        plt.plot(history.history[loss_key], label=loss_label)
    if val_loss_key in history.history:
        plt.plot(history.history[val_loss_key], label=val_loss_label)
    plt.title(f'에포크별 손실 변화 ({title_prefix})')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    # 서브플롯 2: AUC
    plt.subplot(1, 2, 2)
    if auc_key in history.history:
        plt.plot(history.history[auc_key], label='학습 AUC')
    if val_auc_key in history.history:
        plt.plot(history.history[val_auc_key], label='검증 AUC')
    plt.title(f'Loss Over Epochs ({title_prefix})')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.savefig(save_filename)
    plt.show()


# --- Main Script ---

# CSV 파일 로드
file_path = 'combined_dataset_with_multi_labels.csv'
df = pd.read_csv(file_path)

# 상관계수 0.3 이상인 피처 리스트
selected_features = [
    'Flow Duration', 'Fwd IAT Total', 'Bwd IAT Total', 'Fwd IAT Max',
    'Bwd IAT Max', 'RST Flag Count', 'Bwd IAT Mean', 'Fwd IAT Mean',
    'Bwd IAT Min', 'Fwd PSH Flags', 'Fwd IAT Min', 'FIN Flag Count',
    'Fwd IAT Std', 'Packet Length Std', 'Bwd Packet Length Std',
    'Average Packet Size', 'Packet Length Max', 'Bwd IAT Std',
    'Packet Length Mean', 'Bwd Packet Length Max', 'Dst Port'
]

# 데이터프레임에서 피처 선택
df_selected = df[selected_features]

# 로그 변환 적용 (log1p)
df_log_transformed = df_selected.copy()
for feature in selected_features:
    df_log_transformed[feature] = np.log1p(df_log_transformed[feature])

# Label에 따라 정상(0)과 비정상(>=1)으로 분류
df['Is Malicious'] = df['Label'].apply(lambda x: 1 if x >= 1 else 0)
X = df_log_transformed
y = df['Is Malicious']

# 데이터의 샘플링의 사이즈를 지정 
# 테스트 시에 사용 
X_sampled, _, y_sampled, _ = train_test_split(X, y, train_size=0.5, random_state=42)

# 오버샘플링 전 데이터 분포 확인
print("Before Oversampling :")
print(y_sampled.value_counts())

# SMOTE를 사용하여 오버샘플링
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_sampled, y_sampled)

# 오버샘플링 후 데이터 분포 확인
print("\nAfter Oversampling :")
print(pd.Series(y_resampled).value_counts())

# Z-score 정규화 적용 (StandardScaler)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_resampled)

# PCA 적용
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# 설명된 분산 비율 계산
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

# 95% 분산을 설명하는 주성분 수 선택
n_components = np.argmax(cumulative_variance_ratio >= 0.95) + 1
print(f"\nNumber of components to explain 95% variance: {n_components}")

# PCA 재적용 (선택된 주성분 수로)
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X_scaled)

# 학습 데이터와 테스트 데이터 분리 (PCA 적용된 데이터 사용)
# 7:3 분리 
X_train, X_test, y_train, y_test = train_test_split(X_pca, y_resampled, test_size=0.3, random_state=42)

# --- 모델 학습 설정 ---
BATCH_SIZE = 64 # 모든 배치 사이즈를 64로 통일
epochs_teacher = 100
num_samples_teacher = X_train.shape[0]
end_step_teacher = np.ceil(num_samples_teacher / BATCH_SIZE).astype(np.int32) * epochs_teacher

# 교사 모델에 프루닝 적용을 위한 설정
'''
기존 모델은 초기 30% 최대 60%까지 가지치기 비율을 조정했었습니다.
프로젝트 수행 시에는 큰 문제는 발생하지 않았다만은. 
초기부터 30%으로 크게 프루닝을 진행하는 것은 위험할 수 있습니다. 
'''
pruning_params_teacher = {
    'pruning_schedule': tfmot.sparsity.keras.PolynomialDecay( # 점진적 가지치기 적용 
        initial_sparsity=0.00,
        final_sparsity=0.50,
        begin_step=0,
        end_step=end_step_teacher
    )
}

# 교사 모델 정의 (MLP)
input_dim = X_train.shape[1]
teacher_model = Sequential([
    Input(shape=(input_dim,)),
    Dense(50, activation='gelu'), # relu보다 성능은 좋고, 메모리 사용량은 비슷한 gelu사용 
    Dense(50, activation='gelu'),
    Dense(1, activation='sigmoid')
])

# 교사 모델에 프루닝 적용
teacher_model_for_pruning = tfmot.sparsity.keras.prune_low_magnitude(teacher_model, **pruning_params_teacher)

# 교사 모델 컴파일
teacher_model_for_pruning.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss=tf.keras.losses.BinaryCrossentropy(),
    metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
)

# 콜백 설정
early_stopping = EarlyStopping(
    monitor='val_auc',
    patience=10,
    mode='max',
    restore_best_weights=True
)
reduce_lr_teacher = ReduceLROnPlateau(
    monitor='val_loss', # 교사 모델은 val_student_loss가 없으므로 val_loss 사용
    factor=0.2,
    patience=10,
    min_lr=0.0001
)

# 교사 모델 학습 (프루닝 적용)
history_teacher = teacher_model_for_pruning.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=epochs_teacher,
    batch_size=BATCH_SIZE, # 배치 사이즈 32로 변경
    callbacks=[tfmot.sparsity.keras.UpdatePruningStep(), early_stopping, reduce_lr_teacher],
    verbose=1
)

# 교사 모델 평가
y_pred_proba = teacher_model_for_pruning.predict(X_test).flatten()
y_pred = (y_pred_proba >= 0.5).astype(int)
print("\nPruned Teacher Model Classification Report:")
print(classification_report(y_test, y_pred, target_names=['Normal', 'Malicious']))
roc_auc = roc_auc_score(y_test, y_pred_proba)
print(f"Pruned Teacher Model ROC-AUC Score: {roc_auc:.4f}")

# 학습 곡선 시각화 (함수 사용)
plot_training_history(history_teacher, 'Pruned Teacher', 'training_history_pruned_teacher.png')

# 교사 모델 저장 (파일명에 "123" 추가)
teacher_model_for_pruning.save('123pruned_teacher_model.keras')

# Distiller 클래스 정의
class Distiller(Model):
    def __init__(self, student, teacher, alpha=0.1, temperature=5.0):
        super(Distiller, self).__init__()
        self.teacher = teacher
        self.student = student
        self.alpha = alpha
        self.temperature = temperature

    def compile(self, optimizer, metrics, student_loss_fn, distillation_loss_fn):
        super(Distiller, self).compile(optimizer=optimizer, metrics=metrics)
        self.student_loss_fn = student_loss_fn
        self.distillation_loss_fn = distillation_loss_fn

    def train_step(self, data):
        x, y = data
        teacher_output = self.teacher(x, training=False)
        with tf.GradientTape() as tape:
            student_output = self.student(x, training=True)
            soft_teacher = tf.nn.softmax(tf.concat([1 - teacher_output, teacher_output], axis=1) / self.temperature) # 온도 함수를 통하여 분포를 조정 
            soft_student = tf.nn.softmax(tf.concat([1 - student_output, student_output], axis=1) / self.temperature)
            distillation_loss = self.distillation_loss_fn(soft_teacher, soft_student) * (self.temperature ** 2)
            student_loss = self.student_loss_fn(y, student_output)
            loss = self.alpha * student_loss + (1 - self.alpha) * distillation_loss
        trainable_vars = self.student.trainable_variables
        gradients = tape.gradient(loss, trainable_vars)
        self.optimizer.apply_gradients(zip(gradients, trainable_vars))
        self.compiled_metrics.update_state(y, student_output)
        results = {m.name: m.result() for m in self.metrics}
        results.update({"student_loss": student_loss, "distillation_loss": distillation_loss})
        return results

    def test_step(self, data):
        x, y = data
        student_output = self.student(x, training=False)
        student_loss = self.student_loss_fn(y, student_output)
        self.compiled_metrics.update_state(y, student_output)
        results = {m.name: m.result() for m in self.metrics}
        results.update({"student_loss": student_loss})
        return results

    def call(self, inputs):
        return self.student(inputs)

# 학생 모델 정의
student_model = Sequential([
    Input(shape=(input_dim,)),
    Dense(20, activation='gelu'),
    Dense(10, activation='gelu'),
    Dense(1, activation='sigmoid')
])

# Distiller 인스턴스 생성 (프루닝된 교사 모델 사용)
distiller = Distiller(
    student=student_model,
    teacher=teacher_model_for_pruning,
    alpha=0.1,
    temperature= 5 
    '''
    온도 함수는 분포를 다양하게 만들어, 학생모델에서도 다양한 데이터를 학습할 수 있도록 합니다 .
    LLM에서는 온도가 1~2만 되어도 큰 값이지만, 증류의 경우 Teacher모델의 성능이 뛰어난 경우에는  
    특정 상황만 편향되게 학습할 가능성이 있으므로, 성능이 높을 수록 온도를 높게 설정해도 됩니다.
    ''' 
)

# Distiller 컴파일
distiller.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    metrics=['accuracy', tf.keras.metrics.AUC(name='auc')],
    student_loss_fn=tf.keras.losses.BinaryCrossentropy(),
    distillation_loss_fn=tf.keras.losses.KLDivergence(),
)

# 지식 증류 콜백 설정
reduce_lr = ReduceLROnPlateau(
    monitor='val_student_loss',
    factor=0.2,
    patience=10,
    min_lr=0.0001
)

# 학생 모델 학습 (지식 증류, epochs=100)
history_distiller = distiller.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=100,
    batch_size=BATCH_SIZE, # 배치 사이즈 32로 통일
    callbacks=[early_stopping, reduce_lr],
    verbose=1
)

# 학생 모델 평가
student_model = distiller.student
y_pred_proba_student = student_model.predict(X_test).flatten()
y_pred_student = (y_pred_proba_student >= 0.5).astype(int)
print("\nStudent Model Classification Report:")
print(classification_report(y_test, y_pred_student, target_names=['Normal', 'Malicious']))
roc_auc_student = roc_auc_score(y_test, y_pred_proba_student)
print(f"Student Model ROC-AUC Score: {roc_auc_student:.4f}")

# 학습 곡선 시각화 (학생 모델 - 함수 사용)
plot_training_history(history_distiller, 'Student', 'training_history_student_model.png')

# 학생 모델 저장 (파일명에 "123" 추가)
student_model.save('new_student_model_with_distillation.keras')

# 프루닝(Pruning) 적용 (학생 모델)
epochs_pruning = 50
num_samples_pruning = X_train.shape[0]
# end_step_pruning 계산 시 BATCH_SIZE (32) 사용
end_step_pruning = np.ceil(num_samples_pruning / BATCH_SIZE).astype(np.int32) * epochs_pruning

pruning_params_student = {
    'pruning_schedule': tfmot.sparsity.keras.PolynomialDecay(
        initial_sparsity=0.00,
        final_sparsity=0.50,
        begin_step=0,
        end_step=end_step_pruning
    )
}

model_for_pruning = tfmot.sparsity.keras.prune_low_magnitude(student_model, **pruning_params_student)

model_for_pruning.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss=tf.keras.losses.BinaryCrossentropy(),
    metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
)

# Note: reduce_lr 콜백은 'val_student_loss'를 모니터링합니다.
# 프루닝 단계에서는 'val_loss'를 사용해야 하므로, 교사 모델의 reduce_lr_teacher를 재사용하거나 새로 정의합니다.
reduce_lr_pruning = ReduceLROnPlateau(
    monitor='val_loss', # 프루닝 단계에서는 'val_loss' 사용
    factor=0.2,
    patience=10,
    min_lr=0.0001
)

pruning_callbacks = [
    tfmot.sparsity.keras.UpdatePruningStep(),
    early_stopping,
    reduce_lr_pruning # 수정된 콜백 사용
]

history_pruning = model_for_pruning.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=epochs_pruning,
    batch_size=BATCH_SIZE, # 배치 사이즈 32로 통일
    callbacks=pruning_callbacks,
    verbose=1
)

y_pred_proba_pruned = model_for_pruning.predict(X_test).flatten()
y_pred_pruned = (y_pred_proba_pruned >= 0.5).astype(int)
print("\nPruned Student Model Classification Report:")
print(classification_report(y_test, y_pred_pruned, target_names=['Normal', 'Malicious']))
roc_auc_pruned = roc_auc_score(y_test, y_pred_proba_pruned)
print(f"Pruned Student Model ROC-AUC Score: {roc_auc_pruned:.4f}")

# 학습 곡선 시각화 (프루닝된 학생 모델 - 함수 사용)
plot_training_history(history_pruning, 'Pruned Student', 'training_history_pruned_model.png')

# 프루닝된 학생 모델 저장 (파일명에 "123" 추가)
model_for_pruning.save('new_pruned_student_model.keras')

# 프루닝 래퍼 제거 (양자화 준비)
stripped_pruned_model = tfmot.sparsity.keras.strip_pruning(model_for_pruning)
stripped_pruned_model.save('new_stripped_pruned_student_model.keras')

# TFLite 모델 양자화 및 성능 평가 (동적 양자화 제거)
quantization_levels = {
    'int8': tf.int8
}

results = {}

def representative_dataset():
    # 데이터셋 크기를 200으로 유지
    num_samples = min(200, len(X_train))
    for i in range(num_samples):
        yield [X_train[i:i+1].astype(np.float32)]

for level_name, dtype in quantization_levels.items():
    print(f"\nQuantizing pruned model to {level_name}...")
    converter = tf.lite.TFLiteConverter.from_keras_model(stripped_pruned_model)
    converter.optimizations = [tf.lite.Optimize.DEFAULT]
    
    # INT8 양자화 설정
    if dtype == tf.int8:
        converter.representative_dataset = representative_dataset
        converter.target_spec.supported_ops = [
            tf.lite.OpsSet.TFLITE_BUILTINS_INT8
        ]
        converter.inference_input_type = tf.int8  # 입력 타입을 int8로 설정
        converter.inference_output_type = tf.int8 # 출력 타입을 int8로 설정
    else:
        # 다른 양자화 유형(예: float16)을 위한 예비 설정
        converter.target_spec.supported_types = [dtype]
        converter.inference_input_type = dtype
        converter.inference_output_type = dtype

    tflite_model = converter.convert()
    tflite_filename = f'123pruned_student_quantized_{level_name}.tflite' # 파일명에 "123" 추가
    with open(tflite_filename, 'wb') as f:
        f.write(tflite_model)
    print(f"Pruned model quantized and saved as '{tflite_filename}'")
    
    interpreter = tf.lite.Interpreter(model_content=tflite_model)
    interpreter.allocate_tensors()
    input_details = interpreter.get_input_details()
    output_details = interpreter.get_output_details()
    
    y_pred_proba_tflite = []
    
    # TFLite 모델 추론
    for i in range(len(X_test)):
        input_data = X_test[i:i+1].astype(np.float32)
        
        # INT8 입력 처리
        if input_details[0]['dtype'] == np.int8:
            input_scale, input_zero_point = input_details[0]['quantization']
            input_data = (input_data / input_scale + input_zero_point).astype(np.int8)
        
        interpreter.set_tensor(input_details[0]['index'], input_data)
        interpreter.invoke()
        output_data = interpreter.get_tensor(output_details[0]['index'])
        
        # INT8 출력 처리
        if output_details[0]['dtype'] == np.int8:
            output_scale, output_zero_point = output_details[0]['quantization']
            output_data = (output_data.astype(np.float32) - output_zero_point) * output_scale
            
        y_pred_proba_tflite.append(output_data[0])
        
    y_pred_proba_tflite = np.array(y_pred_proba_tflite).flatten()
    y_pred_tflite = (y_pred_proba_tflite >= 0.5).astype(int)
    
    print(f"\nClassification Report for {level_name} Quantized Pruned Model:")
    print(classification_report(y_test, y_pred_tflite, target_names=['Normal', 'Malicious']))
    roc_auc_tflite = roc_auc_score(y_test, y_pred_proba_tflite)
    print(f"ROC-AUC Score for {level_name} Quantized Pruned Model: {roc_auc_tflite:.4f}")
    
    results[level_name] = {
        'roc_auc': roc_auc_tflite,
        'classification_report': classification_report(y_test, y_pred_tflite, output_dict=True)
    }

print("\nSummary of Quantization Results for Pruned Model:")
for level_name, result in results.items():
    print(f"{level_name}: ROC-AUC = {result['roc_auc']:.4f}")

# 스케일러와 PCA 객체 저장 (파일명에 "123" 추가)
scaler_filename = 'new_scaler_with_pca_keras.pkl'
joblib.dump(scaler, scaler_filename)
print(f"Scaler saved as {scaler_filename}")
pca_filename = 'new_pca_with_pca_keras.pkl'
joblib.dump(pca, pca_filename)
print(f"PCA object saved as {pca_filename}")
```
